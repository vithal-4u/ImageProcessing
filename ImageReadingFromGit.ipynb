```python
import requests

def list_github_folder(owner, repo, folder_path, branch="main"):
    url = f"https://api.github.com/repos/{owner}/{repo}/contents/{folder_path}?ref={branch}"
    response = requests.get(url)

    if response.status_code == 200:
        files = response.json()
        for file in files:
            print(f"Name: {file['name']}, Type: {file['type']}, URL: {file['download_url']}")
    else:
        print(f"Error: {response.status_code}, {response.json()}")

# Example Usage
list_github_folder("vithal-4u", "ImageProcessing", "Wallpapers-37")
```

```python
import os
import cv2
import matplotlib.pyplot as plt
import git

def clone_github_repo(repo_url, local_path):
    if not os.path.exists(local_path):
        print(f"Cloning repository from {repo_url}...")
        git.Repo.clone_from(repo_url, local_path)
    else:
        print("Repository already cloned.")

def display_images_from_folder(folder_path):
    if not os.path.exists(folder_path):
        print(f"Folder '{folder_path}' does not exist!")
        return

    image_extensions = (".png", ".jpg", ".jpeg", ".bmp", ".gif")
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(image_extensions)]

    if not image_files:
        print("No images found in the folder!")
        return

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        image = cv2.imread(image_path)

        if image is None:
            print(f"Skipping '{image_file}' - Unable to load")
            continue

        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        plt.imshow(image)
        plt.title(image_file)
        plt.axis("off")
        plt.show()

# Example usage
github_repo_url = "https://github.com/vithal-4u/ImageProcessing.git"  # Replace with actual GitHub repo URL
local_repo_path = "D:\Study_Data\git_hub\github_images_repo"  # Local directory to clone repo
images_folder = os.path.join(local_repo_path, "Sample_images")  # Adjust subfolder if necessary

clone_github_repo(github_repo_url, local_repo_path)
display_images_from_folder(images_folder)
```
