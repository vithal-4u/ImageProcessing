```python
import requests

def list_github_folder(owner, repo, folder_path, branch="main"):
    url = f"https://api.github.com/repos/{owner}/{repo}/contents/{folder_path}?ref={branch}"
    response = requests.get(url)

    if response.status_code == 200:
        files = response.json()
        for file in files:
            print(f"Name: {file['name']}, Type: {file['type']}, URL: {file['download_url']}")
    else:
        print(f"Error: {response.status_code}, {response.json()}")

# Example Usage
list_github_folder("vithal-4u", "ImageProcessing", "Wallpapers-37")
```
